##### Note about Education and Income Features

We included **Education level** and **Income level** as part of the model features because socioeconomic status has been consistently linked to diabetes risk in multiple studies. 

Research shows that individuals with **lower education and income levels** have higher risks of developing type 2 diabetes. This is due to several interconnected factors:

- **Limited access to healthcare services**: People with lower socioeconomic status often have reduced access to quality healthcare, preventive screenings, and diabetes management programs.  
- **Lifestyle factors**: Lower income and education levels are associated with higher rates of obesity, physical inactivity, and unhealthy diets, all of which are risk factors for diabetes.  
- **Health literacy**: Higher education levels typically correlate with better health literacy, which influences diet choices, physical activity, and understanding of disease prevention.  

For example, a study by Robbins et al. (2005) found that **low socioeconomic status was strongly associated with increased diabetes prevalence** in the U.S. population. Similarly, Agardh et al. (2011) conducted a systematic review and concluded that **lower education level is a consistent predictor of higher diabetes risk** across countries and ethnic groups.

Including these features in our model helps improve its predictive accuracy and reflects the real-world context of diabetes risk, beyond purely biological or clinical factors.

**References:**

- Robbins, J. M., Vaccarino, V., Zhang, H., & Kasl, S. V. (2005). Socioeconomic status and type 2 diabetes in African American and non-Hispanic white women and men: Evidence from the Third National Health and Nutrition Examination Survey. *American Journal of Public Health*, 95(8), 1379-1384.  
- Agardh, E., Allebeck, P., Hallqvist, J., Moradi, T., & Sidorchuk, A. (2011). Type 2 diabetes incidence and socio-economic position: a systematic review and meta-analysis. *International Journal of Epidemiology*, 40(3), 804-818.

### Part D.6 : Feature Engineering & Selection (Correlation Threshold)

To reduce multicollinearity, we checked for highly correlated features (correlation > 0.85) using the Pearson correlation matrix. Features with high correlation were to be removed to improve model performance and reduce redundancy.

However, the result showed:
*Highly correlated features (to drop): []*

This indicates that no features had a correlation greater than 0.85 with others. Therefore, *no features were removed at this step*. This is a good sign of low multicollinearity in the dataset, allowing us to retain all features for model training.

````python
# Find highly correlated features
import numpy as np

corr_matrix = df.corr().abs()
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_features = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.85)]

print("Highly correlated features (to drop):",high_corr_features)
df_selected = df.drop(columns=high_corr_features)
````

We conducted feature selection to remove highly correlated features, ensuring that our model would not be affected by multicollinearity.

##### LightGBM

LightGBM (Light Gradient Boosting Machine) is a fast and efficient gradient boosting framework that uses tree-based learning algorithms. It is particularly well-suited for large datasets and high-dimensional features due to its leaf-wise tree growth strategy, which often results in better accuracy compared to level-wise approaches.

The model was trained using random_state=42 for reproducibility.  
LightGBM is known for its speed, efficiency, and accuracy, making it a strong candidate for binary classification tasks such as diabetes prediction.

````python
# 10. Voting Classifier (Ensemble)
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# Define base models
lr = LogisticRegression(max_iter=1000, random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
xgb = XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss', random_state=42)
lgbm = LGBMClassifier(n_estimators=100, random_state=42)

# Create Voting Classifier
voting_clf = VotingClassifier(
    estimators=[
        ('LogReg', lr),
        ('RF', rf),
        ('XGB', xgb),
        ('LGBM', lgbm)
    ],
    voting='soft'
)

# Train the ensemble model
voting_clf.fit(X_train, y_train)
````

##### Voting Classifier (Ensemble of LR, RF, XGB, LGBM)

The Voting Classifier combines multiple individual models to improve overall prediction performance. Here, we used soft voting, which averages the predicted class probabilities from:

- Logistic Regression  
- Random Forest  
- XGBoost  
- LightGBM

This ensemble approach often helps in capturing different strengths of each model and increases robustness. All base models were trained on the same training data with default parameters (and some tuning like setting random_state for reproducibility).

````python
X = df.drop(columns='Diabetes_binary')
y = df['Diabetes_binary']
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
````

````python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train,y_train)
````

````python
# 1. Logistic Regression

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Predict
y_pred_logreg = lr.predict(X_test)
y_proba_logreg = lr.predict_proba(X_test)[:, 1]

# 2. Evaluation Metrics
accuracy = accuracy_score(y_test, y_pred_logreg)
precision = precision_score(y_test, y_pred_logreg)
recall = recall_score(y_test, y_pred_logreg)
f1 = f1_score(y_test, y_pred_logreg)
roc_auc = roc_auc_score(y_test, y_proba_logreg)

# Print metrics
print("Logistic Regression Metrics")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}")
print(f"ROC-AUC:   {roc_auc:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_logreg))

# 3. Confusion Matrix
cm = confusion_matrix(y_test, y_pred_logreg)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
# 4. Save results to DataFrame
results_logreg = {
    'Model': 'Logistic Regression',
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-score': f1,
    'ROC-AUC': roc_auc
}
results_df = pd.DataFrame([results_logreg])
````

### Feature Selection: Drop Highly Correlated Features

In this step, we aim to reduce the problem of multicollinearity in our dataset by removing features that are highly correlated with others. This helps the model learn more effectively and prevents overfitting.

```python
import numpy as np

# Calculate correlation matrix
corr_matrix = df.corr().abs()

# Select the upper triangle of the correlation matrix
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation > 0.9 (threshold)
high_corr_features = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.9)]

print("Highly correlated features (to drop):", high_corr_features)

# Drop highly correlated features
df_selected = df.drop(columns=high_corr_features)
Interpretation:
 • high_corr_features is a list of feature names that are highly correlated (above 0.9) with other features in the dataset.
 • In the output from the Debug Console, we see:
Highly correlated features (to drop): []
This means there are no features with correlation above 0.9, so no features need to be dropped.
 • As a result, the df_selected dataframe will be the same as the original df since no features were removed.

Note for reporting:
 • It’s important to document that no highly correlated features were found in this step.
